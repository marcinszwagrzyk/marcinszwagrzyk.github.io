```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects
fig, ax = plt.subplots()


# Dodanie tekstu
text = ax.text(0.5, 0.5, 'Przykładowy tekst', fontsize=20, ha='center', va='center', color='white')

# Dodanie efektu obwódki do tekstu
text.set_path_effects([
    path_effects.Stroke(linewidth=2, foreground='black'),
    path_effects.Normal()
])
```
